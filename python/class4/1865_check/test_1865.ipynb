```python
N, M, W = 3, 3, 1
edges = [(1, 2, 2), (1, 3, 4), (2, 3, 1), (3, 1, -3)]
```

```python
MAX_TIME = 10 ** 4
MAX_TOT_TIME = MAX_TIME * (M - 1) + 1

def bellman_ford(
    edges: list[tuple[int, int, int]],
    source: int
) -> list[int]:
    dist = [MAX_TOT_TIME] * (N + 1)
    dist[source] = 0

    for _ in range(N+M-1):
        updated = False
        for u, v, w in edges:
            new_dist = dist[u] + w
            if dist[u] != MAX_TOT_TIME and new_dist < dist[v]:
                dist[v] = new_dist
                updated = True
        if not updated:
            break
    
    for u, v, w in edges:
        if dist[u] != MAX_TOT_TIME and dist[u] + w < dist[v]:
            return "YES"

bellman_ford(edges, )
```
